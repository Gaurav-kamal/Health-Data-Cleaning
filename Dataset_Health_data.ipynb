###The "dropna()" returns a new dataset with no null values.

```python
df = pd.read_csv("/content/Health_data.csv")
new_df=df.dropna()
print(new_df.to_string())                         # to_string() - you can see whole data without any skip
```

###The "dropna(inplace=True)" will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

```python
df = pd.read_csv("/content/Health_data.csv")
df.dropna(inplace = True)       #
print(df.to_string())
```

###The fillna() method allows us to replace empty cells with a value

```python
df = pd.read_csv("/content/Health_data.csv")
df.fillna(60,inplace = True)
print(df.to_string())
```

```python
df = pd.read_csv("/content/Health_data.csv")
df['Calories'].fillna(150,inplace = True)
print(df.to_string())
```

```python
df = pd.read_csv("/content/Health_data.csv")
x = df['Calories'].mean()
df['Calories'].fillna(x,inplace = True)
print(df.to_string())
```

```python
df = pd.read_csv("/content/Health_data.csv")
x = df['Calories'].median()
df['Calories'].fillna(x,inplace = True)
print(df.to_string())
```

```python
df = pd.read_csv("/content/Health_data.csv")
x = df['Calories'].mode()[0]
df['Calories'].fillna(x,inplace = True)
print(df.to_string())
```

###Replacing values

```python
df = pd.read_csv("/content/Health_data.csv")
df.loc[7, 'Duration'] = 45
print(df.to_string())
```

```python
df = pd.read_csv("/content/Health_data.csv")
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
print(df.to_string())
```

```python
df = pd.read_csv("/content/Health_data.csv")
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
print(df.to_string())
```

###Removing Duplicates

```python
df = pd.read_csv("/content/Health_data.csv")
print(df.duplicated())
```

```python
df = pd.read_csv("/content/Health_data.csv")
df.drop_duplicates(inplace = True)
print(df.to_string())
```
